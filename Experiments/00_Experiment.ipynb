```python
Log = LogToSTDOUT
Log("Ready to log!")
```

```python
database = Database("00_Experiment.db")
```

```python
database
```

```python
gene = "X1"
FP_OFF={"X1":0,"X2":0,"X3":0} 
FP_ON={"X1":[1,3],"X2":[1,2],"X3":[1,1]} 
```

```python
Log("Constructing query...")
query = HysteresisQuery(database,gene,FP_OFF,FP_ON)
Log("Query constructed!")
```

```python
Log("Performing single gene query...")
searchgraph = query.GeneQuery(0)
searchgraph.matching_label = lambda v : query.matching_label(searchgraph.mgi(v))
Log("Done!")
```

```python
Log("Constructing alignment graph...")
alignment_graph = AlignmentGraph(searchgraph, query.patterngraph, query.matching_relation)
Log("Done!")
```

```python
root_vertex = (0,0)
leaf_vertex = (len(searchgraph.vertices)-1, 4)
```

```python
Log("Traversing alignment graph...")
num_paths = alignment_graph.numberOfPaths(root_vertex, leaf_vertex)
Log("Done!")
```
